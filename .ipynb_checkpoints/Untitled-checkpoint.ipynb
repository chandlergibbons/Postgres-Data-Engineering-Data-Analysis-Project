```python
import matplotlib.pyplot as plt
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:7924@localhost:5432/SQL_HW_1')
conn = engine.connect()
```

```python
pewlett_hackard_emp_db = pd.read_sql("SELECT employees.emp_no, employees.last_name, employees.first_name, titles.title, salary.salary FROM employees LEFT JOIN titles ON employees.emp_title_id = titles.title_id LEFT JOIN salary ON employees.emp_no = salary.emp_no;", conn)
```

```python
pewlett_hackard_emp_db.head()
```

```python
plt.hist(pewlett_hackard_emp_db["salary"])
plt.xlabel("employee numbers")
plt.ylabel("salary range")
plt.title("Histogram of most common salary ranges for employees")
plt.show()
```

```python
avg_salary_per_title = pewlett_hackard_emp_db.groupby(["title"])[["salary"]].mean()

plt.bar(x = avg_salary_per_title.index.values, height = avg_salary_per_title["salary"] )
plt.xlabel("titles")
plt.ylabel("salary")
plt.xticks(rotation = 45)
plt.title("Average salary for each job title")
plt.show()
```
